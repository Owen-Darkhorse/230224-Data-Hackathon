```python
%matplotlib widget

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
```

```python
cyclica = pd.read_csv("230219-cyclicaNoEntry.csv")

zdata = cyclica.coord_Z
xdata = cyclica.coord_X
ydata = cyclica.coord_Y
```

```python
fig = plt.figure(figsize=(9, 6))
ax = plt.axes(projection='3d')

ax.scatter3D(xdata, ydata, zdata, 
    s=cyclica.y_Ligand.map({False: 2, True: 10}),
    alpha = cyclica.y_Ligand.map({False: 0.1, True: 1}), 
    color=cyclica.y_Ligand.map({False:'blue', True: 'orange'}), )

ax.set_title("3D Coordinates of Residues")
ax.set_xlabel('x-Coord')
ax.set_ylabel('y-Coord')
ax.set_zlabel('z-Coord')

plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
ax = plt.axes(projection='3d')

ax.scatter3D(data=cyclica, xs = 'feat_PHI', ys = 'feat_TAU', zs = 'feat_PSI', 
    s=cyclica.y_Ligand.map({False: 2, True: 10}),
    alpha = cyclica.y_Ligand.map({False: 0.2, True: 1}), 
    color=cyclica.y_Ligand.map({False:'blue', True: 'orange'}), )

ax.set_title("3D Visualuization of Bonding Angles")
ax.set_xlabel('PHI')
ax.set_ylabel('TAU')
ax.set_zlabel('PSI')

plt.show()
```
